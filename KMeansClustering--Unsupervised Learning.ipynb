```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
%matplotlib inline 
```

```python
"""
n_samples = number of data samples to categorize, this could be your data
centers = number of categories to divide the data into
random_state = randomizing data's state
cluster_std = defines how measurements for a group are spread out from the mean(average)
                Low std = most of the group memebers are very close to the mean.
                High std = memeners are spread out.

"""
x, y = make_blobs(n_samples=100, centers=5, random_state=23, cluster_std=2)
```

```python
# creating an estimator with defined clusters
estimator = KMeans(5)

#fiting estimator into the dataset
estimator.fit(x)
```

```python
# Predicting the catagory of each datasets
y_means = estimator.predict(x)
```

```python
# Plotting the result
plt.scatter(x[:, 0], x[:, 1], s=50)
plt.scatter(x[:, 0], x[:, 1], c=y_means, s=50, cmap="rainbow")
plt.show()
```
